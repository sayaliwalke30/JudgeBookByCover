#### Images of book cover are downloaded from Amazon.com

```python
# Read the csv file
header_names = ['Amazon ID (ASIN)', 'Filename', 'Image URL', 'Title', 'Author', 'Category ID',
                'Category']
#with codecs.open(Book_train.csv, mode='r', encoding='utf-8', errors='ignore') as f:
df = pd.read_csv('book_train.csv',delimiter=",", header=None, names=header_names)
```

### So there are total 47863 total images and 20 different categories are available

```python
# Function to download image from Amazon.com
def download_image(i):
    filename = df.iloc[i]['Filename']
    category = df.iloc[i]['Category']
    print(category)
    inner_output_dirpath = os.path.join('Desktop\BookDataset', category)
    
    if not os.path.isdir(inner_output_dirpath):
        os.mkdir(inner_output_dirpath)
    output_filepath = os.path.join(inner_output_dirpath, filename)

    url = df.iloc[i]['Image URL']
    if not os.path.isfile(output_filepath):
        downloaded_img = request.urlopen(url)
        f = open(output_filepath, mode='wb')
        f.write(downloaded_img.read())
        downloaded_img.close()
        f.close()

Parallel(n_jobs=-1)(delayed(download_image)(i) for i in trange(len(df)))
```
